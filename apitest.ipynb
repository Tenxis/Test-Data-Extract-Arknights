```python
import requests

url = 'https://prts.wiki/api.php'
# params = {
#     "action": "query",
#     "prop": "revisions",
#     "titles": "阿米娅", 
#     "rvslots": "main",
#     "rvprop": "content",
#     "formatversion": "2",
#     "format": "json"
# }
params = {
    "action": "query",
    "prop": "revisions",
    "titles": "干员一览/干员id", 
    "rvslots": "main",
    "rvprop": "content",
    "formatversion": "2",
    "format": "json"
}

response = requests.get(url, params=params)
data = response.json()
```

```python
data
```

```python
content = data["query"]['pages'][0]['revisions'][0]['slots']['main']['content']
print(content)
```

```python
with open('operator.csv','w') as file:
    file.write(content)
```
